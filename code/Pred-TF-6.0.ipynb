```python
# Время восхода солнца и луны 
import ephem,datetime
sun = ephem.Sun()
moon = ephem.Moon()
raleigh = ephem.Observer()
raleigh.lon, raleigh.lat = "86.10426", "55.4467"
raleigh.horizon = '-0:34'  # USNO standard atmospheric diffraction
raleigh.pressure = 0       # atmospheric refraction parameters

def riseset_sun(date, f):
    raleigh.date = date
    sun.compute(raleigh)
    sr = ephem.localtime(f(sun))    
    return sr

def riseset_moon(date, f):
    raleigh.date = date
    moon.compute(raleigh)
    sr = ephem.localtime(f(moon))    
    return sr



def sunframe(df):
    #Солнце
    df['sunrise'] = df['data_temp'].map(lambda d: riseset_sun(d, raleigh.next_rising)) 
    df['sunset'] = df['data_temp'].map(lambda d: riseset_sun(d, raleigh.next_setting)) 
    
    df['sunrise_h']=df['sunrise'].dt.hour
    df['sunrise_m']=df['sunrise'].dt.minute/60
    
    df['sunset_h']=df['sunset'].dt.hour      
    df['sunset_m']=df['sunset'].dt.minute/60

    
    df['Dsun']=(df['sunset_h'] + df['sunset_m'])-(df['sunrise_h'] + df['sunrise_m'])
    df['sunrise']=df['sunset_h'] + df['sunset_m']
    df['sunset'] =df['sunrise_h'] + df['sunrise_m']
    

    #луна 5% к точности
    df['sunrise_moon'] = df['data_temp'].map(lambda d: riseset_moon(d, raleigh.next_rising)) 
    df['sunset_moon'] = df['data_temp'].map(lambda d: riseset_moon(d, raleigh.next_setting)) 

    df['sunrise_moon_h']=df['sunrise_moon'].dt.hour
    df['sunrise_moon_m']=df['sunrise_moon'].dt.minute/60
    
    df['sunset_moon_h']=df['sunset_moon'].dt.hour      
    df['sunset_moon_m']=df['sunset_moon'].dt.minute/60
    
    df['sunrise_moon']=df['sunrise_moon_h']+df['sunrise_moon_m']
    df['sunset_moon']=df['sunset_moon_h']+df['sunset_moon_m']
    
    df['Dmoon']=df['sunset_moon']-df['sunrise_moon']
    
    
    # Нормализуем
    df['Dsun']=(df['Dsun']-df['Dsun'].min())/(df['Dsun'].max()-df['Dsun'].min())
    df['sunrise']=(df['sunrise']-df['sunrise'].min())/(df['sunrise'].max()-df['sunrise'].min())
    df['sunset']=(df['sunset']-df['sunset'].min())/(df['sunset'].max()-df['sunset'].min())
    df['Dmoon']=(df['Dmoon']-df['Dmoon'].min())/(df['Dmoon'].max()-df['Dmoon'].min())
    df['sunrise_moon']=(df['sunrise_moon']-df['sunrise_moon'].min())/(df['sunrise_moon'].max()-df['sunrise_moon'].min())
    df['sunset_moon']=(df['sunset_moon']-df['sunset_moon'].min())/(df['sunset_moon'].max()-df['sunset_moon'].min())
    
    
    return df

data_ml=sunframe(data_ml)
#data_ml['Dsun'].plot().grid()
```
