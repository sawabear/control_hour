### Это просто загрузчик данных. Он готовит формат пикл.

Можно избежать этого. Но нужно в остальных блокнотах поправить тоже.

```python
import pandas as pd
```

```python
for cz in range(1,3): #Для 1 и 2 ценовой зоны
    xls = pd.ExcelFile(f'../data/load_train-{cz}cz.xlsx')
    print('Загружен файл:', f'load_train-{cz}cz.xlsx')
    for sheet in xls.sheet_names: #Для каждого листа в файле обработаем фрейм и сохраним
        pd.read_excel(xls, sheet, skiprows=5, header=[0,1]).fillna(method='ffill', axis=0).to_pickle(f'../models/data_{sheet}-{cz}cz.pkl')
        print('Сохранен лист', sheet, 'в отдельный файл', f'data_{sheet}-{cz}cz.pkl')
    print('\n')
```
